```python
# Importing The Libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

SeaBorn has some Built-in dataset

```python
# Total bill vs tip dataset
tips = sns.load_dataset('tips')
tips.head()
```

The `tips` dataset is a built-in seaborn dataset containing information about tips given at a restaurant. It includes the following columns:

*   **total_bill**: The total bill amount.
*   **tip**: The tip amount.
*   **sex**: The gender of the payer (Male or Female).
*   **smoker**: Whether the payer is a smoker (Yes or No).
*   **day**: The day of the week (Thur, Fri, Sat, Sun).
*   **time**: The time of day (Lunch or Dinner).
*   **size**: The size of the party.

```python
tips.shape
```

```python
# Setting a theme for the plots
sns.set_theme()
```

```python
#Visualization the Data
sns.relplot(data = tips , x = "total_bill" , y = "tip" ,col="time", hue = "sex",style="sex", size = "size")
```

Load the iris Dataset

```python
iris = sns.load_dataset('iris')
iris.head()
```

The `iris` dataset is another built-in seaborn dataset. It contains measurements in centimeters for the sepal length, sepal width, petal length, and petal width of 150 iris flowers from three different species:

*   **sepal_length**: The length of the sepal.
*   **sepal_width**: The width of the sepal.
*   **petal_length**: The length of the petal.
*   **petal_width**: The width of the petal.
*   **species**: The species of the iris flower (setosa, versicolor, or virginica).

Scatter Plot

```python
sns.scatterplot(data = iris , x = "sepal_length" , y = "petal_length" , hue = "species")
```

Loading the Titanic Dataset

```python
titanic = sns.load_dataset('titanic')
titanic.head()
```

The `titanic` dataset is a built-in seaborn dataset containing information about passengers on the Titanic. It includes the following columns:

*   **survived**: Survival (0 = No, 1 = Yes).
*   **pclass**: Passenger class (1 = 1st, 2 = 2nd, 3 = 3rd).
*   **sex**: Sex of the passenger.
*   **age**: Age in years.
*   **sibsp**: Number of siblings/spouses aboard.
*   **parch**: Number of parents/children aboard.
*   **fare**: Passenger fare.
*   **embarked**: Port of embarkation (C = Cherbourg, Q = Queenstown, S = Southampton).
*   **class**: Passenger class (First, Second, or Third).
*   **who**: Man, woman, or child.
*   **adult\_male**: Whether the passenger is an adult male.
*   **deck**: Deck the passenger was on.
*   **embark\_town**: Town of embarkation.
*   **alive**: Whether the passenger is alive (yes or no).
*   **alone**: Whether the passenger is alone.

Count plot

```python
sns.countplot(data = titanic , x = "sex")
```

Bar chart

```python
sns.barplot(data = titanic , x = "sex" , y = "survived" , hue = "class")
```
