Matplotlib:
* useful for making Plots & Graph

```python
#importing matplotlib
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
#Create Some data by Numpy
a = np.linspace(0,10,100)
b = np.sin(a)
c = np.cos(a)
```

Plot The Data

```python
# sine Wave make
plt.plot(a,b)
plt.show()
```

```python
# cosine Wave make
plt.plot(a,c)
plt.show()
```

```python
# Adding Title, X-axis & y-axis
plt.plot(a,b)
plt.title("Sine Wave")
plt.xlabel("X variable")
plt.ylabel("Y variable")
```

Create a Parabola

```python
x = np.linspace(-5,5,100)
y = x**2
plt.plot(x,y)
plt.show()
```

```python
#Create Diagram by Symbol
plt.plot(x,y,'r+')
plt.show()
```

```python
plt.plot(x,y,'r.')
plt.show()
```

```python
plt.plot(x,y,'r-')
plt.show()
```

```python
plt.plot(x,y,'b--')
plt.show()
```

Create Multiple Plot & Graph into One Plot

```python
plt.plot(a,b,'r-')
plt.plot(a,c,'b--')
plt.plot(a,np.tan(a),'g-')
plt.show()
```

Bar Plot

```python
languages = ["English","French","Spanish","latin","Bangla"]
people = [100,50,150,300,275]
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
ax.bar(languages,people)
plt.xlabel("Languages")
plt.ylabel("Number of People")
plt.title("Bar Plot")
plt.show()
```

Pie Chart

```python
fig1 = plt.figure()
ax1 = fig1.add_axes([0,0,1,1])
ax1.pie(people,labels=languages,autopct='%1.1f%%')
plt.show()
```

Scatter plot

```python
fig2 = plt.figure()
ax2 = fig2.add_axes([0,0,1,1])
ax2.scatter(a,b)
plt.show()
```

3D Scatter plot

```python
x1 = 20*np.random.random(100)
y1 = np.sin(x1)
z1 = np.cos(x1)
fig3 = plt.figure()
ax3 = fig3.add_axes([0,0,1,1],projection='3d')
ax3.scatter(x1,y1,z1, c=x1, cmap = "Blues" )
plt.show()
```
