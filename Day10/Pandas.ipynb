```python
#importing the pandas library
import pandas as pd
import numpy as np
```

Create a Pandas DataFrame by Sklearn Dataset

```python
#importing the Boston house price data
from sklearn.datasets import fetch_california_housing
housing_dataset = fetch_california_housing()
```

```python
type(housing_dataset)
```

```python
housing_dataset
```

```python
# Convert the Dataset to Pandas DataFrame
housing_df = pd.DataFrame(housing_dataset.data, columns=housing_dataset.feature_names)
```

```python
# Showing top 5 DataFrame data
housing_df.head()
```

```python
#show the number of Rows and Colunms
housing_df.shape
```
